```python
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
from sklearn import metrics  
```

```python
df=pd.read_csv('/kaggle/input/mushroom-classification/mushrooms.csv')
```

```python
df.head()
```

```python
df.info()
```

**Factorizing the dataset**

```python
df2= df.apply(lambda x: pd.factorize(x)[0])
```

```python
df2.info()
```

```python
X= df2.values[:, 1:23]
Y= df2.values[:, 0]
```

**Spliting the dataset to %20 test and %80 train**

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 100)
```

**Decision tree model training**

```python
 clf1 = DecisionTreeClassifier(criterion = "gini",
            random_state = 100,max_depth=7, min_samples_leaf=5)
```

```python
clf1.fit(X_train,y_train)
```

```python
y_pred = clf1.predict(X_test)
```

**Accuracy score**

```python
accuracy_score(y_test, y_pred)*100
```

```python
print(classification_report(y_test, y_pred))
```

**Random Forest model training**

```python
clf2= RandomForestClassifier(n_estimators = 100)
```

```python
clf2.fit(X_train, y_train)
```

```python
y_pred2 = clf2.predict(X_test)
```

```python
metrics.accuracy_score(y_test, y_pred2)*100
```

**Accuracy score**


```python
print(classification_report(y_pred2, y_test))
```
